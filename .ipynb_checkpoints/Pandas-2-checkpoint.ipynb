# Merge, join, and concatenate

```python
df = pd.read_csv("dataset/titanic.csv",sep=",",header=0,index_col=None)
df.head()
```

## Concatenate

```python
 df1 = pd.DataFrame({ 'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']},
                      index=[0, 1, 2, 3])

df2 = pd.DataFrame({ 'D': ['A4', 'A5', 'A6', 'A7'],
                     'F': ['B4', 'B5', 'B6', 'B7'],
                     'G': ['C4', 'C5', 'C6', 'C7'],
                     'H': ['D4', 'D5', 'D6', 'D7']},
                      index=[2, 3, 8, 9])
df1
```

```python
df2
```

## along row axis

```python
pd.concat([df1,df2],axis=0,join="outer")
```

## along column axis

```python
pd.concat([df1,df2],axis=1,join="inner")
```

## Append

```python
df1.append(df2,ignore_index='True')
```

```python
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
df1.append(s1, ignore_index=True)
```

# Merging/Joining

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K2', 'K3', 'K4', 'K5'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
    
left
```

```python
pd.merge(left,right,on='key',how="inner",indicator='indicator_column')
```

```python
pd.merge(left,right,on='key',how="outer",indicator='indicator_column')
```

```python
pd.merge(left,right,on="key",how='left',indicator='indicator_column')
```

```python
pd.merge(left,right,on="key",how='right',indicator='indicator_column')
```

## Join

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},
                       index=['K0', 'K2', 'K3'])
```

```python
left.join(right,how="inner")
```

```python
left.join(right,how="outer")
```

# Apply,map,Applymap

## Apply

```python
df_new.apply(np.sum,axis=0)
```

## map
Applied to Series

```python
df["Age"].map(lambda x : x+100)
```

## Applymap
Applied elementwise to all data points

```python
df.applymap(lambda x : str(x)+"1")
```
